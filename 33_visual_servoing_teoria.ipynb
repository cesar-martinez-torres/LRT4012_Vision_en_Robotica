# 34 · Introducción a Visual Servoing (Servo-Visión)

## Concepto general
El visual servoing utiliza información visual directamente como señal de realimentación para controlar el movimiento de un robot.

### Tipos principales
**IBVS (Image-Based Visual Servoing):**
- El error se define en el espacio de la imagen.
- Se buscan características visuales y se compara su posición con la deseada.

**PBVS (Position-Based Visual Servoing):**
- El error se define en el espacio 3D.
- Se estima la pose del objetivo relativa a la cámara.


## Formulación básica del error visual
Sea el vector de características extraídas de la imagen:

$\mathbf{s} = [s_1, s_2, ..., s_n]^T$

Y el vector deseado:

$\mathbf{s}^* = [s_1^*, s_2^*, ..., s_n^*]^T$

El error visual se define como:

$\mathbf{e} = \mathbf{s} - \mathbf{s}^*$


## Ley de control simplificada
Una forma clásica del controlador es:

$\dot{\mathbf{q}} = -\lambda \, L_s^{+} \, \mathbf{e}$

donde:
- $\dot{\mathbf{q}}$ son las velocidades articulares o del robot,
- $\lambda$ es una ganancia positiva,
- $L_s^{+}$ es la pseudoinversa de la matriz de interacción visual,
- $\mathbf{e}$ es el error visual.
